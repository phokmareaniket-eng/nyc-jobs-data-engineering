# Essential Imports

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
import re
```

```python
# ======================================================
#  Spark Session
# ======================================================
def get_spark_session():
    return SparkSession.builder \
        .appName("NYC_Jobs_Analysis") \
        .getOrCreate()

spark = get_spark_session()
```

```python
# ======================================================
# Bronze Layer (Raw Load)
# ======================================================

def load_raw_data(path):
    return spark.read.csv(path, header=True, inferSchema=True)
```

```python
# ======================================================
# Silver Layer (Cleaning + Feature Engineering)
# ======================================================

def clean_columns(df):
    new_columns = []
    for col_name in df.columns:
        clean_name = col_name.lower()
        clean_name = re.sub(r'[ ,;{}()\n\t=/-]+', '_', clean_name)
        clean_name = re.sub(r'__+', '_', clean_name)
        clean_name = clean_name.strip('_')
        new_columns.append(clean_name)

    return df.toDF(*new_columns)


def clean_data(df):

    df = clean_columns(df)

    # Drop unnecessary columns
    cols_to_drop = [
        "recruitment_contact",
        "job_description",
        "to_apply",
        "additional_information",
        "work_location_1"
    ]

    df = df.drop(*cols_to_drop)

    # Convert date columns
    date_cols = ["posting_date", "post_until",
                 "posting_updated", "process_date"]

    for c in date_cols:
        df = df.withColumn(c, to_timestamp(col(c)))

    # Remove duplicates
    df = df.dropDuplicates(["job_id"])

    return df


def feature_engineering(df):

    df = df.withColumn(
        "average_salary",
        (col("salary_range_from") + col("salary_range_to")) / 2
    )

    df = df.withColumn(
        "salary_gap",
        col("salary_range_to") - col("salary_range_from")
    )

    df = df.withColumn(
        "posting_year",
        year(col("posting_date"))
    )

    df = df.withColumn(
        "requires_masters",
        when(lower(col("minimum_qual_requirements"))
             .contains("master"), 1).otherwise(0)
    )

    return df


def create_silver_layer(path):

    df = load_raw_data(path)
    df = clean_data(df)
    df = feature_engineering(df)

    df.cache()

    return df
```

```python
# ======================================================
#  Gold Layer (Separate KPI Tables)
# ======================================================

def kpi_top_categories(df):

    return df.groupBy("job_category") \
        .agg(count("*").alias("category_job_count")) \
        .orderBy(desc("category_job_count"))


def kpi_salary_distribution(df):

    return df.groupBy("job_category").agg(
        avg("average_salary").alias("avg_salary"),
        max("average_salary").alias("max_salary"),
        min("average_salary").alias("min_salary")
    )


def kpi_highest_salary_per_agency(df):

    window = Window.partitionBy("agency") \
                   .orderBy(desc("average_salary"))

    return df.withColumn("rank",
                         row_number().over(window)) \
             .filter(col("rank") == 1) \
             .select(
                 "agency",
                 col("business_title").alias("highest_paid_title"),
                 col("average_salary").alias("highest_agency_salary")
             )


def kpi_avg_salary_last_two_years(df):

    latest_year = df.agg(max("posting_year")).first()[0]

    return df.filter(
        col("posting_year").isin([latest_year, latest_year - 1])
    ).groupBy("agency").agg(
        avg("average_salary").alias("avg_salary_last_2_years")
    )


def kpi_degree_salary_correlation(df):

    corr_value = df.stat.corr(
        "requires_masters",
        "average_salary"
    )

    return spark.createDataFrame(
        [(corr_value,)],
        ["degree_salary_correlation"]
    )


def kpi_highest_paid_skills(df):

    skills = ["python", "java", "data", "engineering", "management"]

    skill_df = df.select(
        "average_salary",
        lower(col("preferred_skills")).alias("preferred_skills")
    )

    result = None

    for skill in skills:
        temp = skill_df.withColumn(
            "skill_flag",
            when(col("preferred_skills").contains(skill), 1).otherwise(0)
        ).groupBy().agg(
            avg(
                when(col("skill_flag") == 1,
                     col("average_salary"))
            ).alias(f"{skill}_avg_salary")
        )

        result = temp if result is None else result.crossJoin(temp)

    return result
```

```python
# ======================================================
#  Save Gold Tables
# ======================================================

def save_gold_tables(df):

    kpi_top_categories(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/top_categories")
    

    kpi_salary_distribution(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/salary_distribution")

    kpi_highest_salary_per_agency(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/highest_salary_per_agency")

    kpi_avg_salary_last_two_years(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/avg_salary_last_2_years")

    kpi_degree_salary_correlation(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/degree_salary_corr")

    kpi_highest_paid_skills(df).coalesce(1) \
        .write.mode("overwrite") \
        .parquet("/dataset/gold/highest_paid_skills")
```

```python
# ======================================================
#  Visualization
# ======================================================
import matplotlib.pyplot as plt

def visualize_top_categories(spark):

    df = spark.read.parquet("/dataset/gold/top_categories")

    pdf = df.limit(10).toPandas()

    plt.figure()
    plt.barh(pdf["job_category"], pdf["category_job_count"])
    plt.xlabel("Number of Jobs")
    plt.ylabel("Job Category")
    plt.title("Top 10 Job Categories")
    plt.tight_layout()
    plt.show()

```

```python
# ======================================================
#  Main Pipeline
# ======================================================

def run_pipeline():

    silver_df = create_silver_layer("/dataset/nyc-jobs.csv")

    # Save silver dataset
    silver_df.write.partitionBy("posting_year").mode("overwrite") \
        .parquet("/dataset/silver/nyc_jobs")

    # Create & save gold KPI tables
    save_gold_tables(silver_df)
    visualize_top_categories(spark)

# Execute
run_pipeline()
print("Job Successful")

```
